# Ejercicio 2



1.   Importarlo como nombre athletes



```python
athletes = pd.read_csv('https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv')
```

```python
athletes.head(12)
```


2.   Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.


```python
%%time
athletes[(athletes["gold"]>0 ) | (athletes["silver"]>0 ) | ((athletes["bronze"]>0 ) ) ]
```



3.   Calcule la masa muscular por fila con una función como



```python
def masa_muscular(row):
  """
  Esta función realiza el calculo de la masa muscular como funcion del peso y la estatura de los atletas



  Parametros Entrada

  row: fila del dataframe


  Retorna

  el calculo de la masa corporal

  """
  return round(((row["weight"])/(row["height"]**2)), 2)
```

```python
athletes['Masa_Muscular'] = athletes.apply(masa_muscular, axis = 1)
```

```python
athletes
```



4.   Agrupe los datos  calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?



```python
weight_mean_men = athletes.groupby(by = ['sport'])[['weight']].mean()
weight_mean_men.sort_values(by = 'weight', ascending=False)
```

```python
print(f"El deporte que tiene los jugadores con mayor peso promedio es: {weight_mean_men.idxmax(axis = 0)[0]}")
```



5.   ¿Cuál es el pais con más participación femenina?



```python
women_plays = athletes[athletes["sex"]=="female"].groupby(by = ['nationality'])[["sex"]].count()
women_plays.sort_values(by = 'sex', ascending=False)
```

```python
print(f"El pais que contiene mayor mujeres participantes es: {women_plays.idxmax(axis = 0)[0]}")
```
